```python
def cash_flow(stock):
    r = requests.get(url=f"https://financialmodelingprep.com/api/v3/cash-flow-statement/{stock}?limit=120&apikey={KEY}")
    income = r.json()
    df = pd.DataFrame.from_dict(income)
    return df

tsla_cash_flow = cash_flow('TSLA')
```

```python
def ttm_ratios(stock):
    r = requests.get(url=f"https://financialmodelingprep.com/api/v3/ratios-ttm/{stock}?apikey={KEY}")
    ratios = r.json()
    df = pd.DataFrame.from_dict(ratios)
    return df
```
